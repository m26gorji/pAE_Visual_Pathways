**Coefficients (c)**: The array **c** holds the Gabor coefficients of the image. These coefficients are a complex representation of the image, where each **coefficient** corresponds to a **certain frequency and location** in the original image. This representation is useful for tasks such as texture analysis, feature extraction, and even compression.


---

**complex128**: This data type in NumPy specifies that each complex number is stored using 128 bits in total:

**64** bits are used for the real part.

**64** bits are used for the imaginary part.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from ltfatpy import gabimagepars, dgtreal, gabdual, idgtreal

# Load an image and convert to grayscale
image_path = '/content/12(1).jpeg'
image_path2 = '/content/12(2).jpeg'
# image_path = '/content/008 (1).jpeg'
# image_path2 = '/content/008 (2).jpeg'



image = Image.open(image_path).convert('L')
image_data = np.array(image, dtype=np.float64)


# Define Gabor filter parameters
M, N = image_data.shape
(a, M, L, N, Ngood) = gabimagepars(M, N, N)
# a (int) – Length of time shift
# M (int) – Number of frequency channels
# L (int) – Length of transform to do
# N (int) – Total number of time steps
# Ngood (int) – Number of time steps (columns in the coefficients matrix) that contain relevant information. The columns from Ngood-1 until N-1 only contain information from a zero-extension of the signal.


# Module of Canonical dual window calculation
g = gabdual('gauss', a, M)
c, Ls, gnum = dgtreal(image_data, g, a, M)
# M: frequency shift
# c (numpy.ndarray) – M*N array of complex128 coefficients.
# Ls (int) – length of input signal
# g (numpy.ndarray) – updated window function. dtype is float64.

# print(np.shape(c)) 128*255*255

# to do: as an analysis, plot a series of reconstructed images for 5, 10, 15, and 20 images are made zero
# c[:5, :, :] = 0
# Reconstructed Image
reconstructed_image = idgtreal(c, gnum, a, M)[0]

# Find the magnitude (for visualization)
magnitude = np.abs(c)

# Display
plt.figure(figsize=(17, 6))
plt.subplot(1, 2, 1)
plt.imshow(image_data, cmap='gray')
plt.title('Original Image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(reconstructed_image, cmap='gray')
plt.title('Reconstructed Image 1 (5 images are made zero)')
plt.axis('off')


# Num_pic = 20
# Num_row_fig = int(Num_pic/5)
# plt.figure(figsize=(18, Num_pic))
# for i in range(1, Num_pic+1):
#   plt.subplot(Num_row_fig, 5, i)
#   plt.imshow(magnitude[i], cmap='gray')
#   plt.title(f"The output of Gabor {i}")
#   plt.axis('off')
```
